## Métodos numéricos básicos

### Método babilónico para la raíz cuadrada

- Problema: ¿cuál es la raíz cuadrada de N?

- La *receta informal* del método babilónico es así:

  1. Suponga que la raíz de $r=N/2$.
  2. A continuación haga esta operación: $(r + N/r )/2$
  3. El resultado de la operación llamelo $r$.
  4. Repita desde el paso 2 hasta que este content@ con la respuesta.

- Aplíque el método para sacar la raíz cuadrada de 10.

#### Solución a mano

- ¿Cuál es la raíz cuadrada de 10?

- Aquí está la solución siguiendo el procedimiento a mano: 

  1. Empezamos con: r = 5
  2. Calculamos: $(r + N/r )/2=(5+10/5)/2=3.5$
  3. Ahora $r = 3.5$.
  4. Repito desde 2.
  2. Calculamos: $(r + N/r )/2=(3.5+10/3.5)/2=3.17857$
  3. Ahora $r = 3.17857$.
  4. Repito desde 2.
  2. Calculamos: $(r + N/r )/2=(3.17857+10/3.17857)/2=3.1623194$
  3. Ahora $r = 3.1623194$.
  4. Repito desde 2.
  2. Calculamos: $(r + N/r )/2=(3.1623194+10/3.1623194)/2=3.16227766$
  3. Ahora $r = 3.16227766$.
  4. Repito desde 2.
  2. Calculamos: $(r + N/r )/2=(3.16227766+10/3.16227766)/2=3.16227766$

### Método de Gauss para hacer integrales:

- Problema: ¿cuál es el resultado de la integral $\int_{-1}^{1} f(x) dx$?

- La receta informal del método de Gauss es así:

  1. Calcule la función $f$ en: $x=-1/\sqrt{3}$
  2. Calcule la función $f$ en: $x=+1/\sqrt{3}$
  3. Sume los dos valores y ese es el resultado aproximado de la integral.

### Método de Euler para resolver ecuaciones diferenciales

- Problema: ¿cuál es la solución a la ecuación diferencial $dy/dx=f(x,y)?$.

- La receta informal del método de Euler es así:

  1. Supongamos que en un punto inicial $xini$ usted conoce cuánto vale $y$ y es $yini$ (condición inicial) y quiere calcular cuánto vale $y$ en un punto más adelante, en $xini+h$. 
  2. Llame $x$ a $xini$ y $y$ a $yini$
  3. Calcule $y+hf(x,y)$.
  4. Llame $y$ a este resultado.
  5. Llame $x$ a $x+h$.
  4. Repita desde 3 cuántas veces desee.
  
- Aplíquelo para resolver la ecuación: $dy/dx=y^2$, suponiendo que $y=1$ en $x=0$ y que $h=0.1$.

#### Solución a mano:

- ¿Cuál es la solución a la ecuación $dy/dx=y^2$, suponiendo que $y=1$ en $x=0$?

- Aquí está la solución siguiendo el procedimiento a mano:

  1. $xini = 0$, $yini=1$, $h = 0.1$.
  2. Llamo $x=xini=0$, $y=yini=1$
  3. Calculo $y+hf(x,y)=1+0.1\times 1 = 1.1$.
  4. Llamo $y=1.1$
  5. Llamo $x=x+h=0+0.1=0.1$
  6. Repita 3
  3. Calculo $y+hf(x,y)=1.1+0.1\times 1.1^2 = 1.22$.
  4. Llamo $y=1.22$
  5. Llamo $x=x+h=0.1+0.1=0.2$
  6. Repita 3
  3. Calculo $y+hf(x,y)=1.2+0.1\times 1.22^2 = 1.3688$.

## Algoritmos

### ¿Qué son?

- Queremos responder ahora a la pregunta de ¿cómo representar de manera más formal las recetas de los métodos vistos?

- A la representación formal de una receta matemática se la conoce como un *algoritmo*. 

- La palabra viene del nombre del matemático árabe de la edad media *Abu Abdallah Muḥammad ibn Mūsā al-Jwārizmī*, para los amigos *al-Juarismi*, que inventó este concepto.

<center><img src="https://upload.wikimedia.org/wikipedia/commons/d/dd/Al-Khwarizmi_portrait.jpg" width=200></center>

- En tiempos más recientes los algorítmos están relacionados con los computadores y se definen como *la secuencia de pasos que permiten, dados unos datos de entrada, realizar operaciones lógicas y matemáticas para obtener un estado o solución final*.

### Pseudolenguaje y pseudocódigo

- Para *formalizar* los algorítmos es necesario acordar una serie de reglas de lenguaje que sean comunes a todos.

- Por ejemplo, cuando se quiere describir en el algoritmo los datos de entrada debe usarse la palabra **ENTRADA** o cuándo se quiere mostrar el resultado del algorítmo se usa la palabra **SALIDA**.

- A ese conjunto de palabras y reglas se las llama **Pseudolenguaje** (porque no es un lenguaje humano y tampoco un lenguaje de computador.

- A los algorítmos escritos usando pseudolenguaje se los llama **pseudocódigos**.

- Abajo se muestra un ejemplo de pseudocódigo:

### Pseudolenguaje usado en Burden

#### Elementos básicos

- Entrada y salida:

  - Todo algorítmo tiene una entrada y una salida.
  - Para describir la entrada use la palabra **ENTRADA**.
    - Aquí también se puede hacer una descripción detallada de cuáles son las entradas del algorítmo.
    - Normalmente aquí se enumeran simplemente los números que sirven de entrada.
    - Cada número de entrada debe recibir un nombre y se lo llama **variable**.
    - Es posible que en algunos algorítmos las entradas no sean números.

  - Para describir la salida se usa la palabra **SALIDA**.
    - La salida debe estar clara desde el principio del algorítmo.
    
    - La salida puede ser de dos tipos:
    
      - Uno o varios números: en este caso en el paso respectivo se usa la palabra **SALIDA** y entre paréntesis el número.  Ejemplo: **SALIDA**(I).
      
      - La salida puede ir acompañada de un texto descriptivo.  En ese caso el texto se coloca entre comillas simples o dobles y si va acompañado de una variable se separa por coma.  Ejemplo: **SALIDA**("La integral es:",I).
    
- Descripción de los pasos: 

  - Cada paso del algorítmo lo describimos con la palabra **Paso**.  Ejmplo: **Paso 1**, **Paso 2**, etc.

- Cálculos:

  - Cuando en un paso de un cálculo se tiene que realizar una operación matemática sencilla se usa la palabra **Determine**.  Ejemplo: **Determine** c = a+b.

  - Es común que el resultado de toda operación matemática se deba almacenar en una variable.

- Variables:

  - Las variables son palabras o letras que se usan para representar un valor de entrada, de sálida o un valor parcial.

  - Cuando se quiere almacenar un valor en una variable se usa el signo "=".  Ejemplo: y = 2 ó tolerancia = 1e-3.
  
  - Las variables en el pseudolenguaje se parecen a las variables de las matemáticas pero no son lo mismo.  Por ejemplo, en pseudolenguaje se puede escribir $x = x + 1$ indicando que se haga la operación $x+1$ y lo que dé se guarde en $x$.  En matemáticas $x = x+1$ conduce una expresión falsa $0=1$.

- Tipos de variables:

  - En una variable se pueden almacenar distintos tipos de valores:

    - **Números**: variables escalares numéricas. Ejemplo: x = 0.1.
    - **Palabras**: cadenas de caracteres.  Ejemplo: nombre = "jorge".
    - **Listas**: también llamadas _variables indexadas_.  Se usan cuando con un mismo nombre se quieren guardar varios valores.  Ejemplo: $x_1 = 0.1$, $x_2 = 0.2$. 
    - **Funciones**: son procedimientos que se supone el que ejecuta el algoritmo sabe hacer.   Ejemplo: $y = tan(x)$.  En este caso la palabra `tan` representa una función, un procedimiento que el algoritmo sabe ejecutar sin que se le explique nada.

##### Un ejemplo de algorítmo con los elementos básicos

- Queremos escribir este método en pseudocódigo:

  - Problema: ¿cuál es el resultado de la integral $\int_{-1}^{1} f(x) dx$?

  - La receta informal del método de Gauss es así:

    1. Calcule la función $f$ en: $x=-1/\sqrt{3}$
    2. Calcule la función $f$ en: $x=+1/\sqrt{3}$
    3. Sume los dos valores y ese es el resultado aproximado de la integral.
    
  - Aplíquelo para resolver la ecuación: $\int_{-1}^{1} \sqrt{|\tan x|} dx$

- Se escribiría así:
  
  - **ENTRADA**: función $f$.
  - **SALIDA**: Valor $I$ de la integral $\int_{-1}^{1} f(x) dx$
  - *Paso 1* **Determine** $x_1 = -1/\sqrt{3}$, $x_2 = -1/\sqrt{3}$ 
  - *Paso 2* **Determine** $I_1=f(x_1)$, $I_2=f(x_2)$

  - *Paso 3* **Determine** $I = I_1 + I_2$
  - *Paso 4* **SALIDA**("La integral es: ",$I$)


#### Ciclos de repetición definida o ciclo para

- Cuando un algoritmo necesita que unos pasos se repitan, es necesario indicarle cuántas veces.  A este tipo de *estructura lógica* se la llama un **ciclo de repetición**.

- Cuando el número de veces que se quiere repetir algo se sabe de antemano hablamos de un **ciclo de repetición definida** o **ciclo para**.

- Todo ciclo for tiene una variable asociada llamada **contador** y una estructura llamada **iterador**.

- Por ejemplo para repetir unos pasos 5 veces se escribe: **Para** $i$ = 1, 2, ..., 5. 
  
  - En este caso el contador es $i$ y se puede usar dentro del ciclo como una variable.

  - En este caso el iterador es 1, 2, ..., 5.  Los iteradores son normalmente listas de valores. 

- Los iteradores pueden ser más complejos:

  - Iterador en reversa: $5, 4, ..., 1$.

  - Iterador cada 2: $1, 3, 5, ..., 10$.

  - Iteradores de cosas: "a", "b", "c", "d".

##### Un ejemplo de algorítmo con ciclos para

- Queremos escribir este método en pseudocódigo:

  - Problema: ¿cuál es la raíz cuadrada de N?

  - La *receta informal* del método babilónico es así:

   1. Suponga que la raíz de $r=N/2$.  
   2. A continuación haga esta operación: $(r + N/r )/2$
   3. El resultado de la operación llamelo $r$.
   4. Repita desde el paso 2 hasta que este content@ con la respuesta.

- Se escribiría así:
  
  - **ENTRADA**: Número al que se le sacará la raíz cuadrada $N$.
  - **SALIDA**: Valor $r$ de la raíz cuadrada de $N$.
  - *Paso 1* **Determine** $r=N/2$ 
  - *Paso 2* **Para** $i$ = 1, 2, ..., 6 **hacer**
    - *Paso 3* **Determine** $r=(r+N/r)/2$
  - *Paso 4* **SALIDA**("La raíz de ",$N$," es: ",$r$)


#### Condicionales

- Es muy común que en los algorítmos haya que tomar decisiones y para eso se usan los condicionales.

- Para los condicionales se usa la *estructura lógica*, **Si**... **entonces**... **Sino**.

- Un ejemplo: **Si** x > 0 **entonces** **Determine** y = log(x) **Sino** SALIDA("No se puede sacar el logarítmo")

##### Un ejemplo de algorítmo con condicionales

- En el algorítmo de método de Gauss puede pasar que al evaluar la función provista en los valores $\pm1/\sqrt{3}$ la función no de un número real (puede dar un número complejo por ejemplo).

- En este caso el algorítmo puede reescribirse cuidandónos de ese caso:
  
  - **ENTRADA**: función $f$.
  - **SALIDA**: Valor $I$ de la integral $\int_{-1}^{1} f(x) dx$
  - *Paso 1* **Determine** $x_1 = -1/\sqrt{3}$, $x_2 = -1/\sqrt{3}$ 
  - *Paso 2* **Si** $f(x_1)$ no es real **entonces**:
     - *Paso 3* **SALIDA**("La función no es real en el punto 1")
     - **PARE**.
  - *Paso 3* **Si** $f(x_2)$ no es real **entonces**:
     - *Paso 4* **SALIDA**("La función no es real en el punto 1")
     - **PARE**.
  - *Paso 5* **Determine** $I_1=f(x_1)$, $I_2=f(x_2)$
  - *Paso 6* **Determine** $I = I_1 + I_2$
  - *Paso 7* **SALIDA**("La integral es: ",$I$)


- Una versión alternativa:

  - **ENTRADA**: función $f$.
  - **SALIDA**: Valor $I$ de la integral $\int_{-1}^{1} f(x) dx$
  - *Paso 1* **Determine** $x_1 = -1/\sqrt{3}$, $x_2 = -1/\sqrt{3}$ 
  - *Paso 2* **Si** $f(x_1)$ no es real ó $f(x_2)$ no es real **entonces**:
     - *Paso 3* **SALIDA**("La función no es real en uno de los puntos")
     - **PARE**.
  - **Sino**:
    - *Paso 3* **Determine** $I_1=f(x_1)$, $I_2=f(x_2)$
    - *Paso 4* **Determine** $I = I_1 + I_2$
    - *Paso 5* **SALIDA**("La integral es: ",$I$)


#### Ciclos de repetición condicionada

- Cuando un algoritmo necesita que unos pasos se repitan, es necesario indicarle cuántas veces.  A este tipo de *estructura lógica* se la llama un **ciclo de repetición**.

- Cuando el número de veces que se quiere repetir algo no se sabe de antemano, sino que depende de una algo que pasa dentro del propio ciclo, hablamos de un **ciclo de repetición indefinida** o **ciclo mientras**.

- Los ciclo mientras no necesita de un **contador** pero necesitan de una línea previa para garantizar que la primera vez que se evalúa la condición del ciclo funcione.

##### Un ejemplo de algorítmo con ciclos mientras

- El mismo algorítmo del método babilónico pero usando ciclos mientras en lugar de ciclos para sería:
  
  - **ENTRADA**: Número al que se le sacará la raíz cuadrada $N$.
  - **SALIDA**: Valor $r$ de la raíz cuadrada de $N$.
  - *Paso 1* **Determine** $r=N/2$ 
  - *Paso 2* **Determine** i = 1 **[LINEA DE GARANTÍA DEL MIENTRAS]**
  - *Paso 3* **Mientras** $i\le 6$ haga los pasos 4-5
    - *Paso 4* **Determine** $r=(r+N/r)/2$
    - *Paso 5* **Determine** i = i + 1 **[LINEA DE CONTINUACIÓN]**
  - *Paso 6* **SALIDA**("La raíz de ",$N$," es: ",$r$)

- Nótese que los pasos 2 y 5 no tienen ninguna función en el método original.  Se usan solamente para que la lógica del ciclo mientras funcione.

#### Condiciones de convergencia

- Un caso especial de condiciones de parada en los ciclos mientras es aquel en el que se evalúa la *convergencia* de un método, es decir, si el método si esta llegando a una respuesta después de las repeticiones.

- Para evaluar una condición de convergencia es necesario agregar una variable especial al algoritmo que se conoce como la *variable de memoria*. 

- Adicionalmente cuando se evalua la convergencia es necesario usar otra variable especial denominada *tolerancia*. La tolerancia indica el grado de precisión que busca el algorítmo.
 
  - La tolerancia puede ser absoluta o relativa.

  - La **tolerancia absoluta** es cuando se da un valor de la precisión en las mismas unidades de la respuesta.  Ejemplo: si el algorítmo calcula una profundidad y la tolerancia es de 1 mm, la tolerancia es absoluta.

  - La **tolerancia relativa** es cuando se da un valor de precisión en porcentaje.  Ejemolo: si el algorítmo calcula un peso y la tolerancia es del 1%, la tolerancia es relativa.

##### Un ejemplo de algorítmo con condición de convergencia


- En la versión del algoritmo del método babilónico abajo se muestra un ejemplo del uso de una variable de memoria:

  - **ENTRADA**: Número al que se le sacará la raíz cuadrada $N$, tolerancia absoluta $TOL$ **[VARIABLE DE TOLERANCIA]**.
  - **SALIDA**: Valor $r$ de la raíz cuadrada de $N$.
  - *Paso 1* **Determine** $r=N/2$ 
  - *Paso 2* **Determine** $r_m$ = 0 **[VARIABLE DE MEMORIA]**
  - *Paso 3* **Mientras** $|r-r_m|\ge TOL$ haga los pasos 4
    - *Paso 4* **Determine** $r_m=r$ **[VARIABLE DE MEMORIA]**
    - *Paso 5* **Determine** $r=(r+N/r)/2$
  - *Paso 6* **SALIDA**("La raíz de ",$N$," es: ",$r$," con error ",TOL)

- Otra vez, aquí los pasos 2 y 4 no estaban en el método original y se usan exclusivamente por que se desea un criterio de ciclo basado en la precisión.


- El mismo algoritmo pero con tolerancia relativa sería:

  - **ENTRADA**: Número al que se le sacará la raíz cuadrada $N$, tolerancia relativa $TOL$ **[VARIABLE DE TOLERANCIA]**.
  - **SALIDA**: Valor $r$ de la raíz cuadrada de $N$.
  - *Paso 1* **Determine** $r=N/2$ 
  - *Paso 2* **Determine** $r_m$ = 0 **[VARIABLE DE MEMORIA]**
  - *Paso 3* **Mientras** $|r-r_m|\ge TOL|r|$ haga los pasos 4
    - *Paso 4* **Determine** $r_m=r$ **[VARIABLE DE MEMORIA]**
    - *Paso 5* **Determine** $r=(r+N/r)/2$
  - *Paso 6* **SALIDA**("La raíz de ",$N$," es: ",$r$," con error de ",TOL,"%")

- Nótese que como ahora $TOL$ es un porcentaje, debe usarse $TOL|r|$ para indicar cuál es el máximo error que se admite.  Ejemplo: si se está calculando una temperatura y el valor actual es $r=50 C$ y la tolerancia es $TOL=1\%=0.01$, entonces el máximo error admitido es $50\times 0.01=0.5 C$.

#### Listas

- En muchos algoritmos es indispensable guardar los resultados de cada paso del método usando una sola variable.  En ese caso hablamos de una lista.

- Las listas se representan en el pseudolenguaje con variables con índice.  Ejemplo: $x_1, x_2, x_3, \cdots$.

##### Un ejemplo de algorítmo con listas

- Queremos escribir este método en pseudocódigo:

  - **SALIDA**: Lista de los valores de $x$ y los valores de $y$.
  - *Paso 3* **Determine** $x_0=xini$, $y_0=yini$.
  - *Paso 4* **Para** $i$ = 0, 1, 2, ..., N-1 **hacer**
    - *Paso 5* **Determine** $y_{i+1}=y_{i}+h f(x_{i},y_{i})$
    - *Paso 6* **Determine** $x_{i+1}=x_{i}+h$
  - *Paso 7* **SALIDA**("Los valores de la función en la lista de puntos ",$x_k$," es la lista: ",$y_k$)

- En el paso 3, que no está en el método original, se dan los valores iniciales de la lista.

- Nótese que en la salida $x_k$ se usa para denotar *toda la lista* (no solo el valor en la posición $k$).
