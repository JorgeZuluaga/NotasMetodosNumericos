### Ejemplos

- Considere las siguientes situaciones:

  - Supongamos que se conoce a cuánto está el dólar hoy, se sabe cuánto está cambiando por día y se quiere predecir a cuánto estará mañana.

  - Una cámara de fotodetección observa un vehículo y mide su velocidad.  La policía quiere sabe en que parte de la autopista estará el vehículo 10 minutos después.

### Poniéndo nombres a las cosas

- En todas estas situaciones hay cosas en común:

1. Hay una cantidad que cambia (el valor del dolar, la posición del vehículo, la temperatura) a medida que otra cantidad aumenta (el tiempo).

  - Al tiempo lo llamamos la **variable independiente**.  A la variable independiente en la mayoría de los problemas se lo representa con la letra $x$. 
  
  - Decimos que la primera cantidad es una **función** del tiempo y escribimos simbólicamente: tasa de cambio del dolar $D(x)$, posición del vehículo $P(x)$, temperatura media de la Tierra $T(x)$.

  - En general una función se representa como $f(x)$.

2. Conocemos el valor de la primera cantidad (la función) en un valor específico de la segunda cantidad (la variable independiente).  A eso se lo llama **condición inicial**.  Por ejemplo conocemos el valor dele dolar hoy.

  - Llamamos $x_0$ al valor de la variable independiente en la que la función es concida, así que sabemos $D(x_0)$, $P(x_0)$, $T(x_0)$.

3. Conocemos el valor de la velocidad de cambio de la función en el valor específico de la variable independiente.

  - Sabemos por el cálculo **la velocidad de cambio de una función es la derivada de la función** que se denota como $df/dx$ o $f'(x)$ (que a su vez es una función, es decir que también cambia).

4. Lo que queremos calcular el valor de la función en un punto futuro $x$ (el valor del dolar mañana, la posición del vehículo en 1 hora, la temperatura en 100 años).

  - En síntesos queremos calcular $f(x)$ conociendo $f(x_0)$ y $f'(x_0)$

## Aproximaciones de Taylor

### Primeras aproximaciones

- La primera aproximación que podemos hacer del valor de la función es que la función en $x$ será igual a la función en $x_0$ (ej. suponemos que el dolar mañana valdrá lo mismo que el dolar hoy):

  $$
  f(x)\approx f(x_0)
  $$

  A esta la llamamos la **aproximación de orden 0**.

- La segunda aproximación que podemos hacer es suponer que el valor de la velocidad de cambio de la función se mantiene constante entre $x_0$ y $x$.  Es nos permite obtener la **aproximación de orden 1**:

  $$
  f(x)\approx f(x_0) + f'(x_0) (x-x_0)
  $$

  image.png


> **RECORDEMOS**: Una de las interpretaciones geométricas de la primera derivada de una función es que esta representa la pendiende de la gráfica de $f(x)$.  La fórmula de la pendiente $m$ de una recta entre dos puntos $(x_1,y_1)$ y $(x_2,y_2)$ es:

  $$
  m = \frac{y_2-y_1}{x_2-x_1}
  $$

### Aproximación de orden 2

- Las siguientes aproximaciones no son tan fáciles de demostrar matemáticamente, pero basta mostrar la **aproximación de orden 2** para conocer las aproximaciones superiores:

  $$
  f(x)\approx f(x_0) + f'(x_0) (x-x_0) + \frac{1}{2!} f''(x_0) (x-x_0)^2
  $$

  Para esta aproximación hemos asumido que además de la primera derivada $f'(x_0)$ conocemos la **segunda derivada** $f''(x_0)$ de la función.

- **RECORDEMOS**: El factorial de un número entero se define como el producto de todos los números desde ese entero hasta 1.  Simbólicamente:

  $$
  n! = n (n-1) (n-2) \cdots 1
  $$ 
  
  Ejemplo: $4! = 4 \cdot 3 \cdot 2 \cdot 1 = 24$

- En Python el factorial está como un algoritmo incluído en la biblioteca `math` de modo que puede calcularse así:

```python
import math
math.factorial(4)
```

Alternativamente se puede usar:

```python
from math import factorial
factorial(4)
```

### Aproximaciones sucesivas

- Una generalización de la aproximación de orden $n$:

  $$
  f(x)\approx f(x_0) + f'(x_0) (x-x_0) + \frac{1}{2!} f''(x_0) (x-x_0)^2 + \frac{1}{3!} f'''(x_0) (x-x_0)^3+\cdots+\frac{1}{n!} f^{(n)}(x_0) (x-x_0)^n
  $$

- Aquí $f^{(n)}(x)$ es la derivada de orden $n$ y suponemos que se puede calcular.

- **RECORDEMOS**: Todas las funciones a las que se les puede encontrar las primeras $n$ derivadas, por lo menos en un intervalor $[a,b]$ pertenecen a un conjunto que en matemáticas se llama $C^n[a,b]$

- No es muy difícil pensar que en realidad el valor exacto de la función se obtiene si sumamos todos los términos hasta *infinito*.

  $$
  f(x) = f(x_0) + f'(x_0) (x-x_0) + \frac{1}{2!} f''(x_0) (x-x_0)^2 + \frac{1}{3!} f'''(x_0) (x-x_0)^3+\cdots+\frac{1}{n!} f^{(n)}(x_0) (x-x_0)^n + \cdots
  $$

## Sucesiones

- A un conjunto de números que obedecen una regla la llamamos una **sucesión**.  Por ejemplo al conjunto $\{1,4,9,16,\ldots\}$ es la sucesión de los cuadrado de los números naturales.

- Al término $n$-esimo de una sucesión al que se lo llama $a_n$ que lo puede representar con una fórmula. En el caso anterior por ejemplo $a_n=n^2$.  

- La sucesión se representa normalmente como $\{a_n\}$. La sucesión anterior se escribe $\{n^2\}$.

- Algunas sucesiones pueden definirse con una fórmula de recurrencia, indicando cómo el valor de los términos de la sucesión se obtiene a partir de los valores anteriores. 

  - La sucesión de los números impares $\{1,3,5,\ldots\}$ se puede definir diciendo que cada término es igual al anterior más 2. Simbólicamente se escribe $a_{n+1} = a_n + 2$.

  - Cuando calculamos la raíz cuadrada usando el método babilónico, la fórmula que escribimos era: $r_{n+1} = (r_n + N/r_n)/2$.  Es decir los distintos valores que obtenemos en esa tarea forman una sucesión.

  - La sucesión de Fibonacci se puede escribir como $a_{n+1} = a_n + a_{n-1}$.

#### Convergencia y divergencia

- Una sucesión puede avanzar hasta *infinito*.  Por ejemplo los términos de la sucesión $\{1,4,9,16,\ldots\}$ son cada vez más grandes.  Se dice que **la sucesión diverge**.

- Los términos de una sucesión pueden aproximarse a un valor definido.  Por ejemplo los términos de la sucesión $\{1,1/2,1/3,1/4,\ldots\}=\{1/n\}$ son cada vez más pequeños y cercanos a 0.  Se dice que **la sucesión converge** a 0.  Se escribe:

  $$
  \lim_{n\rightarrow\infty} a_n = \lim_{n\rightarrow\infty} \frac{1}{n} = 0
  $$

- La sucesión del método babilónico $\{r_{n+1} = (r_n + N/r_n)/2\}$ converge a la raíz cuadrada de N.

  $$
  \lim_{n\rightarrow\infty} r_n = \sqrt{N}
  $$

- Muchos métodos numéricos funcionan con sucesiones y decimos que **el método converge** cuando la sucesión correspondiente converge también.

## Series

- Si se suman los primeros términos de una sucesión se forma una **suma parcial**.

  $$
  s_n = a_1 + a_2 + \ldots + a_n = \sum_{i=1}^{n} a_i
  $$

  - Una suma parcial de la sucesión $\{1,1/2,1/3,1/4,\ldots\}$ es:

    $$
    s_3 = 1 + \frac{1}{2} + \frac{1}{3}
    $$

- El conjunto de todas las posibles sumas parciales de una sucesión forma... *otra sucesión*.  A la sucesión de sumas parciales se la llama una **serie**. Simbólicamente:

  $$
  \left\{\sum_{i=1}^{n} a_i\right\}
  $$

   - Ejemplo: de la sucesión $\{1,4,9,16,\ldots\}$ se puede definir la serie: 

   $$
   \{1,1+4,1+4+9,1+4+9+16,\ldots\}
   $$

#### Convergencia de series:

- Como con las sucesiones, las series pueden converger o diverger.  La serie anterior diverge.

- Cuando una serie converge se escribe:

  $$
  \lim_{n\rightarrow\infty} \sum_{i=1}^{n} a_i\ = S
  $$

  alternativamente se puede escribir:

  $$
  S\equiv \sum_{i=1}^{\infty} a_i
  $$

  - Ejemplo 1: la serie 
  $$
  \left\{\sum_{i=1}^{n} \frac{1}{2^i} \right\}
  $$
   converge a $S = 2$

  - Ejemplo 2: la serie $\{\sum_{i=1}^{n} 1/i\}$ se conoce como la serie geométrica y no converge.


#### Cálculo de series con Python

##### El algoritmo

- El algoritmo para calcular la suma parcial de una sucesión es:

  - **ENTRADA**: Fórmula de la sucesión $a_n$, número de términos de la suma parcial $N$.
  - **SALIDA**: Valor de la suma parcial
  - **Paso 1**: S = 0
  - **Paso 2**: Para i = 1, 2, 3, ..., n haga:
    - **Paso 3**: S = S + a(i)
  - **Paso 4**: **SALIDA**("La suma parcial es",S)

- En este algoritmo a la variable $S$ se la llama un acumulador y sirve para calcular una suma usando un ciclo.

##### El programa en Python

```python
def suma_parcial(a,n):
  S = 0
  for i in range(1,n+1):
    S = S + a(i)
  print("La suma parcial es",S)
```

Ejemplo de serie divergente:

```python
suma_parcial(lambda n:n**2,1)
suma_parcial(lambda n:n**2,10)
suma_parcial(lambda n:n**2,100)
```

Ejemplo de serie convergente:

```python
suma_parcial(lambda n:1/2**n,1)
suma_parcial(lambda n:1/2**n,20)
suma_parcial(lambda n:1/2**n,120)
```

#### Sucesiones y series alternantes

- Cuando el signo de los términos de una sucesión cambia de un término al siguiente, decimos que la **sucesión es alternante**.

  - Ejemplo de sucesión alternante: $\{1,-1/2,1/3,-1/4,\ldots\}$.
  
  - La fórmula de esta sucesión es: $\{(-1)^{n-1}/n\}$.  
  
  - Mire que cuando $n$ es impar (primer, tercer, quinto, etc. términos), $n-1$ es par y el valor de $(-1)^{n-1}$ es positivo y negativo en el resto de casos. 

- La serie correspondiente a una sucesión convergente se llama también **serie alternante**.

  - La serie correspondiente a la sucesión anterior es:

    $$
    S=\sum_{i=1}^{\infty} \frac{(-1)^{i-1}}{i}
    $$

  - Como en esta serie el valor de $n$ ya no aparece se acostumbra escribirla también de la forma:
    
    $$
    S=\sum_{n=1}^{\infty} \frac{(-1)^{n-1}}{n}
    $$
    Se dice que el *índice de la serie es un **índice mudo** (puede llamarse de cualquier manera.


La serie del ejemplo anterior se puede calcular con Python así:

```python
suma_parcial(lambda n:(-1)**(n-1)/n,1)
suma_parcial(lambda n:(-1)**(n-1)/n,10)
suma_parcial(lambda n:(-1)**(n-1)/n,100)
suma_parcial(lambda n:(-1)**(n-1)/n,1000)
```
